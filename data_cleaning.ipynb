#### Aggregating the data 
hourly -> average daily data

```python
import pandas as pd
```

```python
df = pd.read_csv("chicago_tracts_weekly_weather_forecast_07_09_2025.csv")
df
```

The first step is to convert time to datetime to be able to aggregate properly.

```python
df.dtypes
```

```python
df["trimmed_time"] = df["time"].str.slice(0,10)
df["cleaned_time"] = pd.to_datetime(df["trimmed_time"])
df
```

```python
df.dtypes
```

868 tracts $\times$ 7 days each = 6076 rows

```python
df_agg = df.drop(columns=["time","trimmed_time"]).groupby(["cleaned_time","tract_id"]).agg("mean").reset_index()
df_agg
```

```python
len(df_agg["tract_id"].unique())
```

```python
# saving this for further use (see map_creation_agg_data.ipynb)
df_agg.to_csv("cleaned_aggregated_data.csv", index=False)
```
